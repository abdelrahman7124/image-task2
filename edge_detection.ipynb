```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def display_img(img):
    new_img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    plt.imshow(new_img)
    plt.show()
```

```python
img_path = "images/xo.png"
img = cv2.imread(img_path)
```

1- sobel:

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(221)
plt.imshow(gray,cmap="gray")
plt.title("original")
plt.axis("off")

G_blur=cv2.GaussianBlur(gray,(3,3),100)
Sobel_X=cv2.Sobel(G_blur,-1,1,0) 
Sobel_Y=cv2.Sobel(G_blur,-1,0,1)
Sobel_or=cv2.bitwise_or(Sobel_X,Sobel_Y)

plt.subplot(222)
plt.imshow(Sobel_X,cmap="gray")
plt.title("Sobel x")
plt.axis("off")
plt.subplot(223)
plt.imshow(Sobel_Y,cmap="gray")
plt.title("Sobel y")
plt.axis("off")
plt.subplot(224)
plt.imshow(Sobel_or,cmap="gray")
plt.title("final")
plt.axis("off")
```

2-canny filter

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur=cv2.GaussianBlur(gray,(3,3),100)
plt.subplot(221)
plt.imshow(blur,cmap="gray")
plt.title("original")
plt.axis("off")

canny=cv2.Canny(blur,22,80)
plt.subplot(222)
plt.imshow(canny,cmap="gray")
plt.title("canny")
plt.axis("off")
```

3-laplacian:

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(221)
plt.imshow(gray,cmap="gray")
plt.title("original")
plt.axis("off")


laplacian = cv2.Laplacian(gray, cv2.CV_8U)
plt.subplot(222)
plt.imshow(laplacian, cmap="gray")
plt.title("Laplacian")
plt.axis("off")
```
